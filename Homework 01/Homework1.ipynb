```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import cv2
import numpy as np

path1 = "color_template.png"
img = cv2.cvtColor(cv2.imread(path1), cv2.COLOR_BGR2RGB)

img_RGB = np.zeros(img.shape,dtype=np.float32)
img_YIQ = np.zeros(img.shape,dtype=np.float32)

#Task 1.1 - Convert RGB 2 YIQ
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        R = img[i][j][0]; G = img[i][j][1]; B = img[i][j][2]
        img_YIQ[i][j][0] = (0.299 * R) + (0.587 * G) + (0.114 * B)        # Y
        img_YIQ[i][j][1] = (0.5959 * R) - (0.2746 * G) - (0.3213 * B)     # I
        img_YIQ[i][j][2] = (0.2115 * R) - (0.5227 * G) + (0.3312 * B)     # Q

plt.title('Color Template - YIQ')
plt.axis(False)
plt.imshow(img_YIQ.astype(np.uint8))
plt.show()

#Task 1.2 - Convert YIQ to RGB
for i in range(img_YIQ.shape[0]):
    for j in range(img_YIQ.shape[1]):
        Y = img_YIQ[i][j][0]; I = img_YIQ[i][j][1]; Q = img_YIQ[i][j][2]
        img_RGB[i][j][0] = (1 * Y) + (0.956 * I) + (0.619 * Q)     # Y
        img_RGB[i][j][1] = (1 * Y) - (0.272 * I) - (0.647 * Q)     # I
        img_RGB[i][j][2] = (1 * Y) - (1.106 * I) + (1.703 * Q)     # Q


plt.title('Color Template - Converted to RGB')
plt.axis(False)
plt.imshow(img_RGB.astype(np.uint8))
plt.show()
```

```python
from PIL import Image
# Task 2
cat1 = cv2.cvtColor(cv2.imread('cat1.jpg'), cv2.COLOR_BGR2RGB)
cat2 = cv2.cvtColor(cv2.imread('cat2.jpg'), cv2.COLOR_BGR2RGB)
cat3 = cv2.cvtColor(cv2.imread('cat3.jpg'), cv2.COLOR_BGR2RGB)
dog1 = cv2.cvtColor(cv2.imread('dog1.jpg'), cv2.COLOR_BGR2RGB)
```

```python
# Flip vertical
cat1_v = cv2.flip(cat1,flipCode=0)
cat2_v = cv2.flip(cat2,flipCode=0)
cat3_v = cv2.flip(cat3,flipCode=0)
dog1_v = cv2.flip(dog1,flipCode=0)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat1_v)
plt.axis(False)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(cat2_v)
plt.axis(False)
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(cat3_v)
plt.axis(False)
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(dog1_v)
plt.axis(False)
plt.show()
```

```python
# Flip horizontal
cat1_h = cv2.flip(cat1,flipCode=1)
cat2_h = cv2.flip(cat2,flipCode=1)
cat3_h = cv2.flip(cat3,flipCode=1)
dog1_h = cv2.flip(dog1,flipCode=1)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat1_h)
plt.axis(False)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(cat2_h)
plt.axis(False)
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(cat3_h)
plt.axis(False)
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(dog1_h)
plt.axis(False)
plt.show()
```

```python
# Rotate left by 90 degree
cat1_lsq = Image.fromarray(np.rot90(cat1))
cat2_lsq = Image.fromarray(np.rot90(cat2))
cat3_lsq = Image.fromarray(np.rot90(cat3))
dog1_lsq = Image.fromarray(np.rot90(dog1))

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat1_lsq)
plt.axis(False)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(cat2_lsq)
plt.axis(False)
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(cat3_lsq)
plt.axis(False)
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(dog1_lsq)
plt.axis(False)
plt.show()
```

```python
# # Rotate right by 90 degree
cat1_rsq = Image.fromarray(np.rot90(cat1,3))
cat2_rsq = Image.fromarray(np.rot90(cat2,3))
cat3_rsq = Image.fromarray(np.rot90(cat3,3))
dog1_rsq = Image.fromarray(np.rot90(dog1,3))

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat1_rsq)
plt.axis(False)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(cat2_rsq)
plt.axis(False)
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(cat3_rsq)
plt.axis(False)
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(dog1_rsq)
plt.axis(False)
plt.show()
```
